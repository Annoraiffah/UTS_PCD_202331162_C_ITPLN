```python
#202331162_Annora Iffah Diniyah Darma
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#202331162_Annora Iffah Diniyah Darma
img = cv2.imread("aku.jpg")  
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.imshow(img_rgb)
plt.title("GAMBAR ASLI")
plt.axis('off')
plt.show()
```

```python
#202331162_Annora Iffah Diniyah Darma
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))
plt.imshow(gray, cmap='gray')
plt.title("GAMBAR GRAY")
plt.axis('off')
plt.show()
```

```python
#202331162_Annora Iffah Diniyah Darma
cerah = cv2.convertScaleAbs(gray, alpha=1, beta=50)

plt.figure(figsize=(10, 5))
plt.imshow(cerah, cmap='gray')
plt.title("GAMBAR GRAY YANG DIPERCERAH")
plt.axis('off')
plt.show()
```

```python
#202331162_Annora Iffah Diniyah Darma
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
kontras = clahe.apply(gray)

plt.figure(figsize=(10, 5))
plt.imshow(kontras, cmap='gray')
plt.title("GAMBAR GRAY YANG DIPERKONTRAS")
plt.axis('off')
plt.show()
```

```python
#202331162_Annora Iffah Diniyah Darma
Gabung_kontras = clahe.apply(cerah)

plt.figure(figsize=(10, 5))
plt.imshow(Gabung_kontras, cmap='gray')
plt.title("GAMBAR GRAY YANG DIPERCERAH DAN DIPERKONTRAS")
plt.axis('off')
plt.show()
```
